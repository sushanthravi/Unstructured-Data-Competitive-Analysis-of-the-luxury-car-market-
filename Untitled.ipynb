```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the RateBeer top beers page
url = "https://www.ratebeer.com/top-beers"

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the section that contains the top beers
    beer_table = soup.find('table', class_='table table-striped')

    # Lists to store beer details
    beer_names = []
    beer_ratings = []

    # Loop through each row in the table to get beer names and ratings
    for row in beer_table.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        if len(columns) > 1:
            beer_name = columns[1].text.strip()  # Beer name
            beer_rating = columns[2].text.strip()  # Beer rating
            beer_names.append(beer_name)
            beer_ratings.append(beer_rating)

    # Create a DataFrame to store the results
    top_beers_df = pd.DataFrame({
        'Beer Name': beer_names,
        'Rating': beer_ratings
    })

    # Display the DataFrame
    print(top_beers_df)

    # Save to CSV if needed
    top_beers_df.to_csv('top_beers_ratebeer.csv', index=False)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```
